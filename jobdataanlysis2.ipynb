```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
df=pd.read_csv("/Users/navkiratsingh/Documents/coding/job_data_merged_1.csv")
df.head()
```

```python
df.drop('Unnamed: 0',axis='columns',inplace=True)
```

```python
df['Country']=df['Location'].str.split(',').str[-1].str.strip()
```

```python
country=df['Country'].value_counts()
x=country.index
y=country
```

```python
plt.figure(figsize=(30,10))
ax= plt.subplot()
plt.bar(x, y)
plt.setp(ax.get_xticklabels(), rotation=30, ha='right')
plt.show()
```

```python
dept=df['Department'].value_counts().nlargest(15)
x=dept.index
y=dept
plt.figure(figsize=(30,10))
plt.bar(x,y)
```

```python
category=df['Category'].value_counts()
x=category.index
y=category
plt.figure(figsize=(10,10))
plt.bar(x,y)
```

```python
DS=df[df['Category']=='Data Scientist']
UI=df[df['Category']=='UI/UX']
SD=df[df['Category']=='Software Developer']
HR=df[df['Category']=='HR']
Business=df[df['Category']=='Business Analyst']
Cloud=df[df['Category']=='Cloud']
```

```python
def counting_visual(dataframe,column_name):
    plt.figure(figsize=(10,10))
    sns.countplot(x=column_name,data=dataframe)
    plt.show()
```

```python
counting_visual(DS,'Workplace')
```

```python
counting_visual(UI,'Workplace')
```

```python
counting_visual(SD,'Workplace')
```

```python
counting_visual(HR,'Workplace')
```

```python
counting_visual(Business,'Workplace')
```

```python
counting_visual(Cloud,'Workplace')
```
